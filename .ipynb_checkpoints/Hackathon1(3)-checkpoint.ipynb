## 1. Energy calculation and basic statistics

Compute the daily energy in WH per square meter of solar panel. For this purpose you use the datasets reporting the solar irradation measure in Caen and Tours (source https://www.ecad.eu/). The irradiation is measured in W/m2 per day. You will use the formula:

C = E_Sol x 24 x P_cr x f_perf

where  

C is the electricity produced in WH/m2 for a day

E_sol is the daily solar radiation in W/m2 

P_cr is the peak power coefficient, set here to  0.18 (monocristal silicium)

f_perf depends upon the system, set here to 0.75.

Remark:

1 W = 1 J/sec

1 WH  is 1W x 3600sec = 3600J

energy/m2 = E_sol * 24 * 3600 J/m2 = E_sol * 24 WH/m2

```python
#Romain Polet

df = pd.read_csv("Radiation.csv")
dates = pd.to_datetime(df['DATE'], format="%Y%m%d")
E_sol_caen = df['Caen'].values
E_sol_tours = df['Tours'].values


P_cr = 0.18
f_perf = 0.75
C_caen = E_sol_caen * 24 * P_cr * f_perf
C_tours = E_sol_tours * 24 * P_cr * f_perf


plt.plot(dates, C_caen, '-b', label = 'Caen')
plt.plot(dates, C_tours, '-r', label = 'Tours')
plt.xlabel('Dates')
plt.ylabel('Electricity production [$Wh/m^2$]')
plt.title('Electricity produced in Caen and Tours from 1974')
plt.legend()
plt.show()

plt.boxplot([C_caen, C_tours])
plt.xticks([1, 2], ["Caen","Tours"])
plt.ylabel('Electricity produced [$Wh/m^2$]')
plt.title('Boxplot of production in Caen and Tours (with outliners)')
plt.show()


dic = {}
dic['Dates'] = dates
dic['Caen'] = C_caen
dic['Tours'] = C_tours
towns = ['Caen', 'Tours']
df_C = pd.DataFrame(dic)
Q1 = df_C[towns].quantile(0.25)
Q3 = df_C[towns].quantile(0.75)
IQR = Q3-Q1
df_clean = df_C[~((df_C[towns] < (Q1 - 1.5 * IQR)) |(df_C[towns] > (Q3 + 1.5 * IQR))).any(axis=1)]
C_caen_clean = df_clean['Caen'].values
C_tours_clean = df_clean['Tours'].values
dates_clean = df_clean['Dates'].to_numpy()


plt.boxplot([C_caen_clean, C_tours_clean])
plt.xticks([1, 2], ["Caen","Tours"])
plt.ylabel('Electricity produced [$Wh/m^2$]')
plt.title('Boxplot of production in Caen and Tours (without outliners)')
plt.show()

n = 30
textsize = 8
fig = plt.figure('Histogram of Caen production', figsize=(13,5))
n1, bins, p = plt.hist(C_caen_clean, n, density = 1, edgecolor = 'b')
plt.xlabel('Electricity produced in Caen [$Wh/m^2$]')
plt.xticks(bins, fontsize = textsize)
plt.ylabel('Probability')
plt.title('Probability of daily production in Caen')
plt.show()

fig = plt.figure('Histogram of Tours production', figsize=(13,5))
n2, bins2, p = plt.hist(C_tours_clean, n, density = 1, edgecolor = 'b')
plt.xlabel('Electricity produced in Tours [$Wh/m^2$]')
plt.xticks(bins2, fontsize = textsize)
plt.ylabel('Probability')
plt.title('Probability of daily production in Tours')
plt.show()


```

The temporal evolution of the production is quite similar in both cities until 2020. After this year we observe most of the outliners that will be removed for next graphs and computations. The graph peaks out and bottoms out every year. This cyclic behavior is due to higer production during the summer than during the winter. Because of the peak of production in 2020 and 2021 (which is three times higher than previous peaks), we can expect most outliers to occur during these years.

The boxplots show a higher median in Tours than in Caen, likely because Tours is further south than Caen. The interquantile range of Tours is also higher, so we expect the standard deviation to be higher. This is probably because Caen is a city that is near the coast, while Tours is more inland. The first boxplot (with outliers) highlights that there are lots of upper extreme data points (compared to the lower extreme data points).  1) Fit Gamma and normal distributions by log-likelihood maximization to

 3) Compare on the same plot the empirical, the  gamma and normal pdf (the
    empirical pdf is an histogram of frequencies).
    
 4) Why is there 3 parameters in python for the Gamma pdf whereas there

Remark : set floc to -0.001 for the gamma.fit (to avoid troubles in case of null observations)


```python
#Clément & Eduardo

date_april=[]
Caen_april=[]
Tour_april=[]
for i in range (len(dates_clean)) :
    if (str(dates_clean[i])[6] == '4') :
        date_april.append(dates_clean[i])
        Caen_april.append(C_caen_clean[i])
        Tour_april.append(C_tours_clean[i])


#Question 2.1.1

mu_Caen, variance_Cean = norm.fit(Caen_april)

alpha_caen, loc_Caen, beta_inverse_Caen = gamma.fit(Caen_april, floc = -0.001)# on commence à -0.001

mu_Tour, variance_Tour = norm.fit(Tour_april)

alpha_Tour, loc_Tour, beta_inverse_Tour = gamma.fit(Tour_april, floc = -0.001) 

#Question 2.1.2

#source bootstrapping.py from LEPL1109 moodle author: M. Hainaut
def log_likelihood_gamma(data, beta, alpha):
    L = []
    for x in data:
        y = gamma.pdf(x,alpha,loc=-0.001,scale = beta)
        L.append(np.log(y))
    return np.sum(L) 

def log_likelihood_norm(data, variance, mu):
    L = []
    for x in data:
        y =  norm.pdf(x,mu,scale = variance)
        L.append(np.log(y))
    return np.sum(L)   

Caen_log_likelihood_gamma = log_likelihood_gamma(Caen_april,beta_inverse_Caen,alpha_caen)
Caen_log_likelihood_norm = log_likelihood_norm(Caen_april,variance_Cean,mu_Caen)
Tours_log_likelihood_gamma = log_likelihood_gamma(Tour_april,beta_inverse_Tour,alpha_Tour)
Tours_log_likelihood_norm = log_likelihood_norm(Tour_april,variance_Cean,mu_Caen)



fig, ax = plt.subplots(4,1,figsize=(3, 6))
ax[0].set_title("Caen : gamma log-likelihood")
ax[0].axis('off')
Caen_boxed_gamma = ax[0].table(cellText=[[np.round(Caen_log_likelihood_gamma,3)]], loc='center', cellLoc='center')
Caen_boxed_gamma.set_fontsize(20)
Caen_boxed_gamma.scale(1, 3)
ax[1].set_title("Caen : norm log-likelihood")
ax[1].axis('off')
Caen_boxed_norm = ax[1].table(cellText=[[np.round(Caen_log_likelihood_norm,3)]], loc='center', cellLoc='center')
Caen_boxed_norm.set_fontsize(20)
Caen_boxed_norm.scale(1, 3)
ax[2].set_title("Tours : gamma log-likelihood")
ax[2].axis('off')
Tours_boxed_gamma = ax[2].table(cellText=[[np.round(Tours_log_likelihood_gamma,3)]], loc='center', cellLoc='center')
Tours_boxed_gamma.set_fontsize(20)
Tours_boxed_gamma.scale(1, 3)
ax[3].set_title("Tours : norm log-likelihood")
ax[3].axis('off')
Tours_boxed_norm = ax[3].table(cellText=[[np.round(Tours_log_likelihood_norm,3)]], loc='center', cellLoc='center')
Tours_boxed_norm.set_fontsize(20)
Tours_boxed_norm.scale(1, 3)
```

After the computation of the values, we can conclude that the normal distribution is better because in both cases, the result of the log-likelihood is higher with the normal distribution than with the gamma distribution and, more graphically, we can see that the normal distribution (red curve) fits better than the gamma dsitribution (green curve).

The gamma distribution seen in lectures automatically starts at 0 when it could in reality start elsewhere. This fact is taken into account in our python library that allows our program to shift our starting point using the floc as a parameter. It's optionnal though and has a starting point set to 0 as default.

2.2. Check if the average daily production in April is the same in Caen and Tours. Let us recall that the null hypothesis is

$H_0$: $\mu_{Caen} = \mu_{Tours}$.

Take care to comment your conclusions. Are all assumptions required to perform this test sastisfied?

Assumptions :

- Independence: The cities are 250 kilometers apart, supporting the assumption of independence between the production data in Caen and Tours.

- Normality: As we saw above, the normal distribution fits our data quite well.

- Equal Variance: As we will see below, the variances cannot be considered equal.

Conclusion : 
We choose $\alpha$ = 0.05 and decide to perform a Student-T test.
The very low p-value (1.372e-06) from the T-test is under 5% and strongly suggests rejecting the null hypothesis, indicating that the means of daily production in Caen and Tours in April are different. Unfortunately all the assumptions are not met so we can't conclude with certainty.

```python
# Samy 

X1    = sm.DescrStatsW(Caen_april)
X2    = sm.DescrStatsW(Tour_april)
Ttest= sm.CompareMeans(X1,X2)
print(Ttest.summary(usevar='pooled'))
print(Ttest.ttest_ind(usevar='pooled',value=0))


```

2.3. Test the equality of variance of daily production in April at Caen & Tours?
$H_0$: $\sigma_{Caen}=\sigma_{Tours}$.

The alternative hypothesis is the following:

$H_1$: $\sigma_{Caen} \neq \sigma_{Tours}$

Conclusion :
We choose $\alpha$ = 0.05 and decide to perform a Bartlett test.
The very low p-value (0.00774), inferior to 5%, indicates a rejection of the null hypothesis, thus the variances are not equal.
As mentionned earlier, the Student-T test requires variances to be equal, but it is likely not the case, so we cannot rely on the conclusion of the Student-T test.

3.1. Do we observe any trend in the yearly solar production of electricity over the considered period?
To answer this question: 

a. You will compute the average daily production (Wh/m2) during April from 1977 up to 2019 (included).

b. You get a time-series of 44 values for each city. Regress these values on the explanatory
variables X=(Year-1977). Don't forget to add a constant term and analyze results. 

c. Plot on the same graph, the predicted and the observed values.

d. Comment your results! 


```python
#Romain Polet
#a)
date_april_reg=[]
Caen_april_reg=[]
Tours_april_reg=[]
days = 0
index = 0
caen_average = 0.0
tours_average = 0.0
for i in range (len(date_april)) :
    if (1977 <= int(str(date_april[i])[0:4]) <= 2019 ):
        caen_average += Caen_april[i]
        tours_average += Tour_april[i]
        days += 1
        if (int(str(date_april[i+1])[3]) != int(str(date_april[i])[3])) :
            Caen_april_reg.append(caen_average / days)
            Tours_april_reg.append(tours_average / days)
            date_april_reg.append(int(str(date_april[i])[0:4]))
            days = 0
            caen_average = 0.0
            tours_average = 0.0
            #index +=1

date_april_reg = np.array(date_april_reg)
Caen_april_reg = np.array(Caen_april_reg)
Tours_april_reg = np.array(Tours_april_reg)

X = date_april_reg-1977
Y_Caen = Caen_april_reg
Y_Tours = Tours_april_reg 

#Clément Deckers
#Inspired by RegressionSpring.py from LEPL1109 moodle. Author: dasha
#b)
Xm       = sm2.add_constant(X)

results_for_Caen = sm2.OLS(Y_Caen,Xm).fit()

results_for_Tours = sm2.OLS(Y_Tours,Xm).fit()
print("Caen:")
print(results_for_Caen.summary())
print("Tours:")
print(results_for_Tours.summary())

observed_values_for_Caen = Y_Caen
observed_values_for_Tours = Y_Tours
predicted_values_for_Caen = results_for_Caen.predict()
predicted_values_for_Tours = results_for_Tours.predict()
```

Although the statistical errors (variance) of our independent terms are much lower than the coefficients (25.136 for Tours and 20.555 for Caen), those of the slopes are much higher relatively to the slope coefficient (especially for Tours), which explains the poor prediction of Y values. It basically means that the independent terms are a lot more reliable than the slopes.

Another thing we have to notice is the difference between the F-statistics. 5.060 for Caen and 0.5176 for Tours. Again we can see that the F-statistic of Tours is 10 times lower than Caen's. Therefore we can suppose that the assumption of a linear evolution of the observed values over the years is false.

Thus, we must be critical of the results observed. Besides, we note that the R-squared values of our models are very low (0.11 for Caen and 0,012 for Tours). The relevance of a model being proportional to its factor, this means that only 1.2% of Y variability is explained by the model of Tours which is clearly not enough to ensure the reliability of this model.

```python
#Clément Deckers
#plot regression 
#c)

plt.figure(figsize=(8, 6))
plt.scatter(X+1977, observed_values_for_Caen, color='blue', label='Observed values')
plt.plot(X+1977, predicted_values_for_Caen, color='red', linewidth=2, label='Predicted values')
plt.xlabel('Years')
plt.ylabel('Average electricity production in Caen [$Wh/m^2$]')
plt.title('Observed vs. Predicted values in Caen')
plt.legend()
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(X+1977, observed_values_for_Tours, color='blue', label='Observed values')
plt.plot(X+1977, predicted_values_for_Tours, color='red', linewidth=2, label='Predicted values')
plt.xlabel('Years')
plt.ylabel('Average electricity production in Tours [$Wh/m^2$]')
plt.title('Observed vs. Predicted values in Tours')
plt.legend()
plt.show()
```

The main observation we can make is that the slope of the linear regression in Caen is steeper than in Tours. This implies a greater rise in solar radiation for Caen. Besides, both slopes are positive. It means either that the solar radiation in both cities is increasing over time or that there is an improvement of solar panels efficiency.

3.2. You want to design a model to forecast the solar electric production for the next day (location Caen only). You will work with data over the period 1977 to 2019. 

Let us denote by C(t) the production on day 't'. The model that we want to fit is called autoregressive and is defined as follows:

$$C(t) = \sum_{k=1}^{10} a_k C(t-k) $$

This model is common in time-series analysis and predicts the production of the next day with the  recent observations.

a. Split the dataset into a training set (1977 to 2010 included) and a validation set (2011 to 2019 included).

b.	Estimate this model with statsmodels on the training set. 

c.	How would you judge the quality of the predictive model? (Analyze statistics reported by statsmodel)

d.	Compute the Mean Absolute Error (MAE) between predicted and real consumptions (on the training set).

```python
# Alexis Thieltgen
#a)
training_set = []
validation_set = []

validation_dates = []

training_counter = 0
validation_counter = 0
for i,date in enumerate(dates_clean):
	
	if 1977 <= int(str(date)[:4]) and int(str(date)[:4]) <= 2010:
		if training_counter < 10:
			training_counter += 1
			continue
		previous_values = [k for k in C_caen_clean[i-10:i+1]]
		training_set.append(previous_values)

	if 2011 <= int(str(date)[:4]) and int(str(date)[:4]) <= 2019:
		if validation_counter < 10:
			validation_counter += 1
			continue
		validation_dates.append(date)
		previous_values = [k for k in C_caen_clean[i-10:i+1]]
		validation_set.append(previous_values)
#b)
training_set = np.array(training_set)
Xt = training_set[:,:10]
Yt = training_set[:,10]

validation_set = np.array(validation_set)
Xv = validation_set[:,:10]
Yv = validation_set[:,10]
#c
results = sm2.OLS(Yt,Xt).fit() #Source : Inspired by RegressionSpring.py from LEPL1109 moodle. Author: dasha
print(results.summary())

```

First, we see that the coefficients are high compared to their respective standard deviation, which means that the coefficient values are accurate and reliable.

Moreover, the $R^2$ value (and the adjusted one) is equal to 89.3%. Therefore, the model (with its 10 independent variables $C(t-1)$ to $C(t-10)$) explains a high proportion of the variance of the dependent variable $C(t)$. It is a relevant indicator of the goodness of fit.

Finally, we see that the $F$ statistic is equal to 1.023e+04, which is significantly higher than $F_{k,n-(k+1),1-\alpha}=F_{10,12274,95\%}=1.831$ ($\alpha=5\%$), and that the p-value is equal to $0 < 5\%$. These datas indicate a strong linear relationship between the independent variables and the dependent variable. Also, this value of $F$ represents the explained variance on the unexplained variance. Our high value of $F$ is therefore a good thing.
